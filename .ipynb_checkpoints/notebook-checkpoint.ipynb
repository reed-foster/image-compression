```julia
using Plots
using LaTeXStrings
using LinearAlgebra
using FFTW
using Images
using TestImages
using Wavelets
```

```julia
function threshold2D!(x, compressionRatio)
    dimensions = size(x)
    (compressionRatio > 1) && (compressionRatio = 1)
    num_threshold = round(Int, length(x)*compressionRatio)
    ind = sortperm(vec(x), alg=QuickSort, by=abs)
    for i = 1:length(x) - num_threshold
        zero_idx = ind[i]
        row = rem(zero_idx-1, size(x)[1])+1
        col = div(zero_idx-1, size(x)[1])+1
        x[row, col] = 0
    end
end
```

# JPEG Image Compression With DCT (Standard Quantization and Hard Thresholding)

```julia
function makeQuantizationTable(qualityFactor)
    # base (Q = 50) quantization table
    Tb = [
        16    11    10    16    24    40    51    61;
        12    12    14    19    26    58    60    55;
        14    13    16    24    40    57    69    56;
        14    17    22    29    51    87    80    62;
        18    22    37    56    68   109   103    77;
        24    35    55    64    81   104   113    92;
        49    64    78    87   103   121   120   101;
        72    92    95    98   112   100   103    99
    ]
    if (qualityFactor < 50)
        S = 5000/qualityFactor
    else
        S = 200 - 2*qualityFactor
    end
    Ts = floor.((Tb.*S .+ 50) ./ 100)
    Ts[Ts .== 0] .= 1;
    return Ts
end

function jpegQuantCompress(imgArray, qualityFactor)
    quantImg = zeros(size(imgArray))
    nnzOrig = 0
    nnzQuant = 0
    quant = makeQuantizationTable(qualityFactor)
    tileSize = 8 # forced to 8 because of quantization table size
    # will chop off excess image if image side-length isn't a multiple of tileSize
    for i = 1:div(size(imgArray)[1],tileSize)
        for j = 1:div(size(imgArray)[2],tileSize)
            slice = view(imgArray, (i-1)*tileSize+1:i*tileSize, (j-1)*tileSize+1:j*tileSize)
            dctSlice = FFTW.dct!(slice.*255);
            # count nonzero DCT elements to get compression ratio
            # this needs to happend here because DCT/IDCT are done in place
            nnzOrig += count(i->(i!=0), dctSlice);
            # needs to be round, not floor/ceil
            # https://farid.berkeley.edu/downloads/publications/tr18.pdf
            quantizedDct = quant.*(round.(dctSlice./quant));
            nnzQuant += count(i->(i!=0), quantizedDct);
            quantizedSlice = FFTW.idct!(quantizedDct)./255;
            quantImg[(i-1)*tileSize+1:i*tileSize, (j-1)*tileSize+1:j*tileSize] = quantizedSlice;
        end
    end
    return nnzQuant/nnzOrig, quantImg
end

function jpegThreshCompress(imgArray, tileSize, compRatio)
    threshImg = zeros(size(imgArray))
    nnzOrig = 0
    nnzQuant = 0
    # will chop off excess image if image side-length isn't a multiple of tileSize
    for i = 1:div(size(imgArray)[1],tileSize)
        for j = 1:div(size(imgArray)[2],tileSize)
            slice = view(imgArray, (i-1)*tileSize+1:i*tileSize, (j-1)*tileSize+1:j*tileSize)
            dctSlice = FFTW.dct!(slice);
            # count nonzero DCT elements to get compression ratio
            # this needs to happend here because DCT/IDCT are done in place
            nnzOrig += count(i->(i!=0), dctSlice);
            threshold2D!(dctSlice, compRatio);
            nnzQuant += count(i->(i!=0), dctSlice);
            thresholdSlice = FFTW.idct!(dctSlice);
            threshImg[(i-1)*tileSize+1:i*tileSize, (j-1)*tileSize+1:j*tileSize] = thresholdSlice;
        end
    end
    return nnzQuant/nnzOrig, threshImg
end

img = Gray.(testimage("airplaneF16"))

# standard JPEG (1992)
rawImg = convert(Array{Float32}, img)
display(Gray.(rawImg))
cr, quantImg = jpegQuantCompress(rawImg, 10)
display(Gray.(quantImg))
print("compression ratio: ", round(cr*1000)/1000)

# hard thresholding
rawImg = convert(Array{Float32}, img)
cr, threshImg = jpegThreshCompress(rawImg, 512, 0.046)
display(Gray.(threshImg))
print("compression ratio: ", round(cr*1000)/1000)
```

# JPEG2000 Compression With Discrete Wavelet Transform

```julia
function waveletCompress(imgArray, wt, compRatio)
    dwt!(rawImg, wt)
    nnzOrig = count(i->(i!=0), rawImg);
    threshold2D!(rawImg, compRatio)
    nnzThresh = count(i->(i!=0), rawImg);
    idwt!(rawImg, wt)
    return nnzThresh/nnzOrig
end

rawImg = convert(Array{Float32}, img)
wt = wavelet(WT.cdf97, WT.Lifting)
cr = waveletCompress(rawImg, wt, 0.50)
display(Gray.(dwtCompressed))
print("compression ratio: ", round(cr*1000)/1000)
```
